```python
import pandas as pd
df = pd.read_csv('all_states_weather_data.csv')
```

```python
df
```

```python
# Convert 'Arrival_Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')

# Creating month and year columns
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

print(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Step 1: Calculate min and max for Temperature
temp_agg = df.groupby(['State', 'Month', 'Year'])['Temperature'].agg(['min', 'max'])

# Step 2: Calculate mean or other metrics for Humidity and Rainfall (as needed)
other_agg = df.groupby(['State', 'Month', 'Year'])[['Humidity', 'Rainfall']].mean()

# Step 3: Combine the results
min_max_temp = temp_agg.join(other_agg).reset_index()

print(min_max_temp.head(20))
```

```python
min_max_temp.rename(columns={'min': 'Min_Temp', 'max': 'Max_Temp'}, inplace=True)
```
